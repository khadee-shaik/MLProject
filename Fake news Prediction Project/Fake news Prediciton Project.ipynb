# Stemming 
# stemming is process to reduce with root node like tier trees....
# acting,actor,actrees - > act

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
    stemmed_content = re.sub('[˄a-zA-Z]',' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')] 
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_data['content'] = news_data['content'].apply(stemming)
```

```python
print(news_data['content'])
```

```python
# speperating the data and labels
x =  news_data['content'].values
y = news_data['Label'].values
```

```python
# converting textual data into numerical data
vectorizer = TfidfVectorizer()
vectorizer.fit(x)
```

```python
X = vectorizer.transform(x)
# print(X)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y,test_size = 0.2,stratify=y,random_state=2)
```

```python
# training the model with logistic regression

model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model_pred = model.predict(x_train)
model_pred
```

```python
acc_model_train = accuracy_score(model_pred,y_train)
print("our train model prediction accuracy score is :- ",acc_model_train)
```

```python
# for tetsing data our accuracy score is :- 
# model.fit(x_test,y_test)   # if this was similar error else similar error
model_pred = model.predict(x_test)
acc_model_test = accuracy_score(model_pred,y_test)
print("our test model prediction accuracy score is :- ",acc_model_test)
```
