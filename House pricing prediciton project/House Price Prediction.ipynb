```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor  
from sklearn import metrics
# !pip install xgboost
```

```python
house_pricing = pd.read_csv("HousingData.csv")
house_pricing
```

```python
house_pricing.shape
```

```python
house_pricing.isnull().sum()
```

```python
house_pricing.describe()
```

#correlation between one varibale to other variables 
we have two types :- 
positive correlation 
negitive correlation

```python
correlation = house_pricing.corr()
```

```python
#construct heapmap to better understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Reds')
```

```python
#splittting the data
x = house_pricing.drop(['MEDV'],axis=1)
y = house_pricing['MEDV']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape , x_test.shape , y_train.shape ,y_test.shape,
```

### model_training
XGBoosting regressor

```python
#loading the model
model = XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

### Evaluation
prediction on training data

```python
pred_train_data = model.predict(x_train)
pred_train_data
```

#compare the train predicition data with original data 
# R squared error   (or) mean absoulate error

```python
score_1 = metrics.r2_score(y_train,pred_train_data)
score_1
```

```python
score_2 = metrics.mean_absolute_error(y_train,pred_train_data)
score_2
```

# visualize the training data on 

# predicting data on test data
# actual MEDV and predicted MEDV

```python
plt.scatter(y_train,pred_train_data)
plt.xlabel("Actual MEDV")
plt.ylabel("Predicted MEDV")
plt.title("Actual MEDV VS Predicted MEDV")
plt.show()
```

```python
pred_test_data = model.predict(x_test)
pred_test_data
```

```python
score_1 = metrics.r2_score(y_test,pred_test_data)
score_2 = metrics.mean_absolute_error(y_test,pred_test_data)

score_1,score_2
```
